Brainstorm List of Ideas:
* Game where you stay on the bottom of the screen as objects fall from the top and you have to catch them. There will be more adverse mechanics.
* This game would help with peoples reaction and pattern recognition speed and skills. It would also improve their hand eye coordination with their fingers as many people arent extremely proficient at moving their fingeres in specific ways.
* Strategy
* Roguelike
* Rhythm

Need: To improve pattern recognition and reaction time speed and skill. Also helps people who may not have a great sense of rhythm improve it.

Problem Statement: Many children and even adults tend to struggle with reaction time and pattern recognition speed. A game that has you read the pattern of falling objects and move to catch them would help improve both of these skills in people of ages above 12. These objects will be synchronised to music to help with the factor of entertainment and keeping people engaged with the game.

Skill Development: To develop the skills to create this game, I will use some of the skills from your course as well as make use of online tutorials to create the mechanics and playfield of my game.

Inputs: It will use the mouse, as well as the arrow keys and enter for level selection.  During play, it will use the left and right arrows/a and d for left/right movement and will use either shift key for a speed boost. There will be a point value assigned to each falling object that will be added and there will also be a percentage value of the objects you have hit.

Processing: The game will constantly be checking for if there has been a collision between the player hitbox and the object hitbox, and if there is one it will ad the score of the object to your total score. If the object reaches the ground then it won't give you any score and it will subtract from your percentage.

Outputs: The game will display: player score and amount of objects hit in a row, player accuracy (percentage), player position, objects and their position. It will show a graphic on screen when you miss an object. There will be music and when you collect an object it will play a sound and the objects will be placed so the sounds line up with the music if you get them correctly. You will also see an overview of your score after completing the level, which will have your accuracy, score and your object judgements.

Transmission: There won't be any online transmission of anything in the game.

Storage: The game will store users scores so you can look at the result screen again. It will also store any settings that you use.

User Interaction: You will navigate the menu using either your mouse or the arrow keys and select things in the menu using left click or enter. After you are in the level you have chosen, you will use the left and right arrows or a and d to move your character to collect the objects and you will use left/right shift to use your characters speed.

Core Gameplay: You will be stuck on the bottom of the screen only able to move left and right using a and d or left/right arrows and increse your speed by holoding left/right shift. When in level select you will be able to use the arrow keys or mouse wheel to navigate the menu and left click or enter to select a level.

Scoring and Feedback: There will be a scoring, accuracy and combo system, as well as feedback for collecting/not collecting the objects. The scoring system will assign a point value to every object and will add that point value to your score when you collect the object. The accuracy metric will give you a percentage of how many objects you have hit during the current level. The combo system will track how many objects you have hit after the previous miss and will reset to 0 if you miss an object. There will also be a graphic that shows when you miss a note, as well as there being sounds that play when you hit/miss, with the hit sound being good and synchronising with the music playing while the miss sound will be almost distracting.

Level Progression: The levels will start super easy without requiring the use of the speed mechanic, but as they go on the levels will require more and more proficiency in the game.

Saving and Loading Data: The game will save your scores on levels by taking a screenshot of the results screen and storing it then letting you access it through the menu.

Performance Requirements: The game should load in under 10 seconds and the controls should feel like they aren't lagging behind the users inputs.

Usability Requirements: There will be a tutorial that shows how the game works that is simple and uses slower objects to give the player a chance to test the game.

Compatibility Requirements: The game should be able to run on pc and it would be nice to also have a browser version, however there won't be any mobile or console compatibility.

Scalability Requirements: The UI is designed in a way where more levels can be added to the level list easily and the levels shouldn't be very big files so storage and performance shouldn't be greatly impacted.

Security Requirements: There won't be any online transmission of anything and so it is alright to have the users scores be stored in a secure folder in the game files.

Reliability and Availability: This game has no online transmission or elements.


-Consideration of Social and Ethical Issues:


Accessibility:
My project will be usable by most people, except for people who are blind or who are unable to move any parts of their body. On top of this deaf people may enjoy the game less as they miss out on the music aspect of the game. The game is so simple that there isnt a need for any colourblindness or anything of the sort, as there aren't any colours necessary to see in the first place.

Privacy and Data Protection:
The game will collect the users scores as scoresheets and allow them to access it through the game at any time, however they won't be accessible through the secure files without admin access.

Fairness and Representation:
My game doesn't include any scenarios that could include any form of bias or stereotyping.

Mental and Emotional Well-Being:
The game wouldn't affect users mental health except for extreme edge cases, like if someone spends a long time trying to get a score or beat a level.

Cultural Sensitivities:
None of the content could be offensive to different cultures.
